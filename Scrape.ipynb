```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import math
```

```python
def getIndusText(index):
    URL = "https://www.indus.epigraphica.de/icitview.php"
    page = requests.post(URL, auth=('icit', 'seal123'), data ={'idicit': str(index)})
    soup = BeautifulSoup(page.content, "html.parser")
    results = soup.find_all("table")
    if(len(results) > 1):
        enc = pd.read_html(results[1].prettify())[0]
        enc = enc.dropna(subset=['text'])
        return enc
    return pd.DataFrame()
    
```
